```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import confusion_matrix,roc_auc_score,roc_curve,classification_report
```

```python
def build_model(algorithm,type_):
    algorithm.fit(X_train,y_train)
    y_pred=algorithm.predict(X_test)
    pl.figure(figsize=(15,10))
    print(algorithm)
    pl.subplot(221)
    pl.title('Confusion Matrix',fontsize=15)
    sns.heatmap(confusion_matrix(y_test,y_pred),fmt='d',annot=True,linewidth=1,cmap='Blues')
    train_scores=algorithm.score(X_train,y_train)
    test_score=algorithm.score(X_test,y_test)
    scores=pd.DataFrame({'Type':[],'Scores':[]})
    scores['Type']=['Train_Score','Test_Score']
    scores['Scores']=[train_scores,test_score]
    
    
    pl.subplot(224)
    ax=sns.barplot(scores['Scores'],scores['Type'])
    for i,j in enumerate('Accuracy: '+round(scores['Scores'],5).astype(str)+' %'):
        ax.text(0.1,i,j,fontsize=20)
    
    if type_=='coef':
        features=algorithm.coef_[0]
        features=pd.DataFrame(features)
        features['Features']=X_train.columns
        features=features.rename(columns={0:'Value'})
        features=features.sort_values('Value',ascending=False)
        pl.subplot(223)
        ax=sns.barplot(features['Value'],features['Features'])
        pl.title('Feature Importances')
        pl.xticks(rotation=90)
        for i,j in enumerate(features['Value']):
            ax.text(0,i,j)
        
    if type_=='feat':
        features=algorithm.feature_importances_
        features=pd.DataFrame(features)
        features['Features']=X_train.columns
        features=features.rename(columns={0:'Value'})
        features=features.sort_values('Value',ascending=False)
        pl.subplot(223)
        ax=sns.barplot(features['Value'],features['Features'])
        pl.title('Feature Importances')
        pl.xticks(rotation=90)
        for i,j in enumerate(features['Value']):
            ax.text(0,i,j)
    
    probablities=algorithm.predict_proba(X_test)[:,1]
    fpr,tpr,thresolds=roc_curve(y_test,probablities)
    pl.subplot(222)
    pl.plot(fpr,tpr,label='Area Under Curve')
    pl.plot([0,1],[0,1],linestyle=':',color='k')
    pl.title('ROC Curve',fontsize=15)
    
    pl.legend()

    pl.show()
    
    
#     sns.heatmap(classification_report(y_test,y_pred))
    print('\n Classification Report \n')
    print(classification_report(y_test,y_pred))
    
logistic=LogisticRegression()
build_model(logistic,'coef')
```

```python
build_model(RandomForestClassifier(),'feat')
```

```python
build_model(DecisionTreeClassifier(),'feat')
```

```python
build_model(SVC(),'none')
```

```python
build_model(PassiveAggressiveClassifier(),'none')
```

```python
build_model(AdaBoostClassifier(),'feat')
```

```python
build_model(GaussianNB(),'none')
```

```python
build_model(GradientBoostingClassifier(),'feat')
```
