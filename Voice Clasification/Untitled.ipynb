```python
import numpy as np
import pylab as pl
import networkx as nx
```

```python
points=[(0,1),(1,5),(5,6),(1,2),(2,3),(2,7)]
```

```python
fig,ax=pl.subplots(2,3,figsize=(10,8))
for i,axes in enumerate(ax.ravel()):
    G=nx.Graph()
    G.add_edges_from(points)
    pos=nx.spring_layout(G,k=.5,center=points[2])
    nx.draw_networkx_nodes(G,pos,node_color='g',ax=axes)
    nx.draw_networkx_edges(G,pos,edge_color='b',ax=axes)
    nx.draw_networkx_labels(G,pos,ax=axes)
    pl.show()
```
