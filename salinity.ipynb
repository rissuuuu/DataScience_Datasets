```python
from sklearn.linear_model import LinearRegression
```

```python
X=a.iloc[:,:-1].values
y=a.iloc[:,-1].values
```

```python
reg=LinearRegression()
reg.fit(X,y)
```

```python
pl.figure(figsize=(10,8))
a=np.arange(32.6,35)
pl.scatter(X,y)
pl.plot(a,reg.coef_*a+reg.intercept_,c='r')
```

```python
reg.score(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pol=PolynomialFeatures(degree=3)
X_pol=pol.fit_transform(X)
pol.fit(X_pol,y)
reg=LinearRegression()
reg.fit(X_pol,y)
```

```python
reg.score(X_pol,y)
```

```python
x_grid = np.arange(min(X), max(X), 0.1)
x_grid = x_grid.reshape(-1,1)
pl.scatter(X,y,s=65)
pl.plot(x_grid,reg.predict(pol.fit_transform(x_grid)) , color='red', linewidth = '6')
pl.xlabel('Slt',fontsize=25)
pl.ylabel('Temp',fontsize=25)
pl.title('salt degerlerine gore temp tahmin gosterimi',fontsize=25)
pl.show()
```

```python
nulls=bottle.isnull().sum().sort_values(ascending=False)
prcnt=round(nulls/len(bottle)*100,2)
df_null=pd.DataFrame(columns=['Attr','Total','Percent'])
df_null.Attr=nulls.index
df_null.Total=nulls.values
df_null.Percent=prcnt.values
df_null
```

```python
df=bottle.copy()
for i in df_null.Attr[df_null['Percent']>30]:
    df=df.drop(i,axis=1)
```

```python
df.head()
```

```python
for i in df.columns:
    if df[i].isnull().sum()>0:
        df[i].fillna(df[i].mode().mean(),inplace=True)
```

```python
sns.heatmap(df.isnull())
```

```python
nulls=df.isnull().sum().sort_values(ascending=False)
prcnt=round(nulls/len(bottle)*100,2)
df_null=pd.DataFrame(columns=['Attr','Total','Percent'])
df_null.Attr=nulls.index
df_null.Total=nulls.values
df_null.Percent=prcnt.values
df_null
```

```python
drop_cols = ['Cst_Cnt', 'Btl_Cnt', 'Sta_ID', 'Depth_ID', 'Depthm']
df_norm = df.drop(drop_cols, axis=1)  #  data for normalization
df_scale = df_norm.copy(deep=True)    
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit_transform(df_scale)
```

```python
for i in df_norm.columns.tolist():
    if df_norm[i].mode()[0]==0.0:
        print(' - ',i,df_norm[i].mode()[0])
        df_norm=df_norm.drop(i,axis=1)
```

```python
df['Salnty'].mode()
df_corr=pd.DataFrame(columns=['Attributes','Correlation'])
df_corr.Attributes=df_norm.corr()['Salnty'].sort_values(ascending=False).index
df_corr.Correlation=df_norm.corr()['Salnty'].sort_values(ascending=False).values

```

```python
df_corr.head()
```

```python
plot_attr = ['R_DYNHT', 'R_SIGMA', 'R_Depth', 'RecInd', 'NH3q',  'T_prec', 'T_degC', 'R_POTEMP', 'O2ml_L']
plot_attr
```

```python
for i in plot_attr:
    if plot_attr[0]==i:
        df_plot=df_corr[df_corr.Attributes==i]
    else:
        df_plot=df_plot.append(df_corr[df_corr.Attributes==i])
df_plot
```

```python
bottle['Year']=((bottle['Depth_ID'].str.split('-',expand=True)[0])+(bottle['Depth_ID'].str.split('-',expand=True)[1])).map(lambda x:str(x)[:4])
bottle['Month']=bottle['Depth_ID'].str.split('-',expand=True)[1].map(lambda x: x[2:4])
```
