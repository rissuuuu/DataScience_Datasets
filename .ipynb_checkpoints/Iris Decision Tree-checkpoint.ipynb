```python
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_graphviz
from graphviz import Source
from IPython.display import display
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
data=sns.load_dataset('iris')
```

```python
data.head()
```

```python
data.shape
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
dtc.score(X_train,y_train),dtc.score(X_test,y_test)
```

```python
graph=Source(tree.export_graphviz(dtc,
                                     out_file=None,
                                     rounded=True,
                                     proportion=False,
                                     feature_names=X.columns,
                                     precision=2,
                                     class_names=['setosa','versicolor','virginica'],
                                     filled=True))
display(graph)
```

```python
dtc=RandomForestClassifier()
dtc.fit(X_train,y_train)
dtc.score(X_train,y_train),dtc.score(X_test,y_test)
```

```python
est_tree=dtc.estimators_[99]
```

```python
dtc.estimators_[:10]
```

```python
est_tree
```

```python
graph=Source(tree.export_graphviz(est_tree,
                                     out_file=None,
                                     rounded=True,
                                     proportion=False,
                                     feature_names=X.columns,
                                     precision=2,
                                     class_names=['setosa','versicolor','virginica'],
                                 ))
display(graph)
```

```python
data['species'][data['species']=='setosa']=0
data['species'][data['species']=='versicolor']=1
data['species'][data['species']=='virginica']=2
```
