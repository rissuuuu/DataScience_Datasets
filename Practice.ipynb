```python
revenue=np.random.randint(10000,20000,20)/1.2
expenses=np.random.randint(10000,20000,20)/1.2
```

```python
profit=[]
for i in range(0,len(revenue)):
    profit.append(revenue[i]-expenses[i])
```

```python
profit
```

```python
tax=[]
for i in profit:
    tax.append(0.3*i)
```

```python
tax=np.round(tax,2)
```

```python
profit_after_tax=list([])
for i in range(0,len(profit)):
    profit_after_tax.append(profit[i]-tax[i])
```

```python
profit_after_tax
```

```python
profit_margin=list([])

for i in range(0,len(profit)):
    profit_margin.append(profit_after_tax[i]/revenue[i])
```

```python
profit_margin=[round((i*100),2) for i in profit_margin]
```

```python
profit_margin
```

```python
a=np.mean(profit_after_tax)
```

```python
good_months=list([])
for i in range(0,len(profit)):
    good_months.append(profit_after_tax[i]>a)
```

```python
good_months
```

```python
bad_months=list([])
for i in range(0,len(profit)):
    bad_months.append(profit_after_tax[i]<a)
```

```python
bad_months
```

```python
pl.plot(profit_after_tax)
pl.plot(np.arange(20),[np.mean(profit_after_tax)]*20,label="mean")
pl.legend()
```

```python
best_month=list([])
for i in range(0,len(profit)):
    best_month.append(profit_after_tax[i]==max(profit_after_tax))
```

```python
best_month
```

```python
worst_month=list([])
for i in range(0,len(profit)):
    worst_month.append(profit_after_tax[i]==min(profit_after_tax))
```
