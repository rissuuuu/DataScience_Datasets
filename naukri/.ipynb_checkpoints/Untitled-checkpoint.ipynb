```python
temp=data[data['Job Title']==' Data Scientist']
temp=temp.reset_index(drop=True)
```

```python
temp=temp['Key Skills']
```

```python
skills={}
for i in range(len(temp)):
    lists=temp[i].split('|')
    for i in range(len(lists)):
        skills[lists[i]]=0
        
```

```python
for i in range(len(temp)):
    lists=temp[i].split('|')
    for i in range(len(lists)):
        skills[lists[i]]+=1
```

```python
a=pd.DataFrame(skills.items())
```

```python
a.head()
```

```python
a=a.groupby(0)[1].sum().sort_values(ascending=False).reset_index()
```

```python
a.reset_index()
```

```python
sns.barplot(data=a.head(6),x=0,y=1)
pl.xlabel("Skills")
pl.ylabel("Counts")
```
